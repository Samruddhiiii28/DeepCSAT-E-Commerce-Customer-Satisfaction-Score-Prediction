## ***6. Feature Engineering & Data Pre-processing***

### 1. Handling Missing Values

````python
# Handling Missing Values & Missing Value Imputation

#Numerical columns->fill with median
df['Item_price']=df['Item_price'].fillna(df['Item_price'].median())
df['connected_handling_time']=df['connected_handling_time'].fillna(df['connected_handling_time'].median())

#categorical columns-> fill with mode
df['Customer_City']=df['Customer_City'].fillna(df['Customer_City'].mode()[0])
df['Product_category']=df['Product_category'].fillna(df['Product_category'].mode()[0])
df['channel_name']=df['channel_name'].fillna(df['channel_name'].mode()[0])

#Text column-> fill with empty string
df['Customer Remarks']=df['Customer Remarks'].fillna('')
````

#### What all missing value imputation techniques have you used and why did you use those techniques?

Median imputation was used for numerical columns such as Item Price and Connected Handling Time because it is robust to outliers. Mode imputation was applied to categorical variables such as Customer City and Product Category because these variables represent categories. For textual data like customer Remarks, missing values were replaced with an empty string that text preprocessing can be applied later.

### 2. Handling Outliers

````python
# Handling Outliers & Outlier treatments
Q1=df['Item_price'].quantile(0.25)
Q3=df['Item_price'].quantile(0.75)

IQR=Q3-Q1

lower_bound=Q1-1.5*IQR
upper_bound=Q3+1.5*IQR

df=df[(df['Item_price']>=lower_bound) & (df['Item_price']<=upper_bound)]
````

##### What all outlier treatment techniques have you used and why did you use those techniques?

The Interquartile Range metod was used to detect and remove outliers from numerical columns such as Item Price. The IQR method is effective because it identifies extreme values that fall far outside the normal data distribution. Removing outliers help improve model accuracy and prevents extreme values from biasing the machine learning model.

### 3. Categorical Encoding

````python
# Encode your categorical columns
from sklearn.preprocessing import LabelEncoder

# Create a label encoder object
le= LabelEncoder()

categorical_cols = ['Customer_City', 'Product_category', 'channel_name', 'category', 'Sub-category', 'Agent_name', 'Supervisor', 'Manager', 'Tenure Bucket', 'Agent Shift']

# Apply label encoding to each categorical column
for col in categorical_cols:
    df[col] = le.fit_transform(df[col].astype(str))
````

#### What all categorical encoding techniques have you used & why did you use those techniques?

Label Encoding was used to convert categorical variables into numerical format.Machine learning algorithm require numerical inputs, so categorical features such as channel name, product category, and agent shift were encoded using Label Encoding. This technique assign a unique integer value to each category while maintaining computational efficiency.

## ***7. ML Model Implementation***

### ML Model - 1 :Linear Regression

````python
# ML Model - 1 Implementation
from sklearn.linear_model import LinearRegression

lr_model=LinearRegression()

lr_model.fit(X_train,y_train)

y_pred_lr=lr_model.predict(X_test)
````

#### 1. Explain the ML Model used and it's performance using Evaluation metric Score Chart.

````python
# Visualizing evaluation Metric Score chart
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
import numpy as np

mse=mean_squared_error(y_test,y_pred_lr)
rmse=np.sqrt(mse)
mae=mean_absolute_error(y_test,y_pred_lr)
r2=r2_score(y_test,y_pred_lr)

print("Mean Squared Error:", mse)
print("Root Mean Squared Error:", rmse)
print("Mean Absolute Error:", mae)
print("R-squared:", r2)
````

````python
#visualizing evaluation Score chart
metrics=['MAE','MSE','RMSE','R-squared']
scores=[mae,mse,rmse,r2]

plt.figure(figsize=(10,6))
plt.bar(metrics,scores)
plt.title('Linear Regression Evaluation Metrics')
plt.show()
````

Linear Regression is a supervised machine learning algorithm used to predict continuous values. Model performance is evaluated using MAE, RMSE and R2 Score.

#### 2. Cross- Validation & Hyperparameter Tuning

````python
# ML Model - 1 Implementation with hyperparameter optimization techniques (i.e., GridSearch CV, RandomSearch CV, Bayesian Optimization etc.)

# ML Model - 1 Implementation with hyperparameter optimization

from sklearn.model_selection import GridSearchCV

param_grid = {
    'fit_intercept':[True, False],
    'copy_X':[True, False]
}

grid_search = GridSearchCV(
    LinearRegression(),
    param_grid,
    cv=5,
    scoring='r2'
)

grid_search.fit(X_train, y_train)

best_model = grid_search.best_estimator_

# Predict using tuned model
y_pred_tuned = best_model.predict(X_test)
````

````python
mse_tuned = mean_squared_error(y_test, y_pred_tuned)
rmse_tuned = np.sqrt(mse_tuned)
mae_tuned = mean_absolute_error(y_test, y_pred_tuned)
r2_tuned = r2_score(y_test, y_pred_tuned)

print("Tuned RMSE:", rmse_tuned)
print("Tuned MAE:", mae_tuned)
print("Tuned R2:", r2_tuned)
````

````python
metrics = ['MAE','RMSE','R2']

before = [mae, rmse, r2]
after = [mae_tuned, rmse_tuned, r2_tuned]

x = np.arange(len(metrics))

plt.figure(figsize=(8,5))

plt.bar(x-0.2, before, width=0.4, label="Before Tuning")
plt.bar(x+0.2, after, width=0.4, label="After Tuning")

plt.xticks(x, metrics)
plt.title("Model Performance Before vs After Hyperparameter Tuning")
plt.legend()

plt.show()
````

##### Which hyperparameter optimization technique have you used and why?

GridSearchCV was used for hyperparameter tuning. It tests multiple parameter combinations using cross validations and selects the best performing model.

##### Have you seen any improvement? Note down the improvement with updates Evaluation metric Score Chart.

Yes, after hyperparamter tuning the model showed slightly improved performance, with lower error values and a higher R2 score compared to the intial model.
